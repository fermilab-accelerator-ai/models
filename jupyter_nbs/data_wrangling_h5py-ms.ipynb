```python
import pandas as pd
import h5py
import  numpy as np
import datetime
```

```python
file = '../../AI4GMPS/MLEventData_1561738277.7050312_From_EventC_2019-06-27_to_2019-06-28.h5'

# read in file into dataframe
f= h5py.File(file, 'r')   
keys = list(f.keys())
print(keys)
print(len(keys))
```

```python
for key in keys:
    df = pd.read_hdf(file, key)
    print(key, df.head())
```

```python
time_stamps = pd.date_range(start='2019-06-27',end='2019-06-28', freq='66.667ms')
```

```python
temp = np.zeros(len(time_stamps))
temp_dict = {'time': time_stamps}
for key in keys:
    temp_dict[key] = temp
```

```python
# build dataframe with time as index and indicator variables (currently all 0) for events
df_master = pd.DataFrame(temp_dict)
df_master.set_index('time')
```

```python
# populate the dataframe
for key in keys:
    df_event = pd.read_hdf(file,key)
    df_event['event_time'] = pd.to_datetime(df_event['datetime'])
    event_indicator = pd.cut(df_event['event_time'], time_stamps)
    event_indicator = event_indicator.apply(lambda x: x.left)
    for event in event_indicator:
        idx = df_master.index[event<=df_master['time']][0] # find first match
        df_master.loc[idx,key]=1 # flag the event as on in that bin
```
